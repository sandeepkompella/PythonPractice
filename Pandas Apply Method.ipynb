```python
import pandas as pd
import numpy as np
stores_sales = pd.read_csv("Movies/Data Files/store_sales.csv")
stores_sales
```

```python
JanAvg = stores_sales["Jan"].agg("mean")
JanAvg
```

```python
stores_sales.apply("mean")
# Mean sales by City
stores_sales.groupby("city").agg("mean")
#mean sales by Stores so that that the average is calculated by row and not column
stores_sales.apply("mean", axis = 1) # axis 0 is used for column wise aggregation
# to restrict mean to only int and float and fix any errors because of strings
stores_sales.select_dtypes(include = ['int64', 'float64']).apply("mean") 
```

```python
#write a function to display if the stores are eligible for a bonus when they have sales > 10
def bonus_func(sales_col):
   elig_values = np.where(sales_col>10, "eligible", "Not eligible")
   return elig_values
stores_sales["Jan_bonus"] = bonus_func(stores_sales["Jan"])
#stores_sales["Jan_bonus"]
stores_sales
```

```python
#add two data frames next to each other
newdf = pd.concat([stores_sales,stores_sales["Jan_bonus"]], axis = 1)
newdf
```
